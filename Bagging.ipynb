```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
# df['Gender']=enc.fit_transform(df.Gender,'Married','Education','Dependents','Self_Employed','Property_Area','Loan_Status')
df['Gender']=enc.fit_transform(df.Gender)
df['Married']=enc.fit_transform(df.Married)
df['Education']=enc.fit_transform(df.Education)
df['Dependents']=enc.fit_transform(df.Dependents)
df['Self_Employed']=enc.fit_transform(df.Self_Employed)
df['Property_Area']=enc.fit_transform(df.Property_Area)
df['Loan_Status']=enc.fit_transform(df.Loan_Status)
```

```python
x=df.drop(columns=['Loan_ID','Loan_Status']).values
y=df['Loan_Status'].values
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sta = StandardScaler()
x_tr=sta.fit_transform(x_tr)
x_te=sta.fit_transform(x_te)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
rfc= RandomForestClassifier(criterion='entropy',random_state=42)
rfc.fit(x_tr,y_tr)
```

```python
y_ptr=rfc.predict(x_tr)
print(classification_report(y_tr, y_ptr))
```

```python
y_pte=rfc.predict(x_te)
print(classification_report(y_te, y_pte))
```
