```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import plotly as py
import plotly.graph_objs as go

from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Mall_Customers.csv')
df
```

```python
features = ['Annual Income (k$)', 'Spending Score (1-100)']
x=df[features]
```

```python
x_s = StandardScaler().fit_transform(x)
```

```python
model = DBSCAN(eps=0.5 , min_samples=5)
clusters = model.fit_predict(x_s)
clusters
```

```python
centroids = pd.DataFrame(x_s).groupby(clusters).mean().values
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x['Annual Income (k$)'], x['Spending Score (1-100)'], c=clusters , cmap='viridis')
plt.title('DBSCAN')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
