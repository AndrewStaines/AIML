```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import plotly as py
import plotly.graph_objs as go

from sklearn.cluster import KMeans
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Mall_Customers.csv')
df
```

```python
x= df.iloc[:,[3,4]].values
x
```

```python
cen =[]
for i in range(1,11):
  kmeans = KMeans(n_clusters=i , init='k-means++', random_state = 42)
  kmeans.fit(x)
  cen.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),cen)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('cen')
plt.show()
```

```python
kmeansmodel = KMeans(n_clusters= 5 , init='k-means++', random_state=42)
y_kmeans = kmeansmodel.fit_predict(x)
```
