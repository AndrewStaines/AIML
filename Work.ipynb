```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Iris.csv')
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
df['Species']=enc.fit_transform(df.Species)
y_tr = df["Species"]
df.drop(columns=['Species', 'Id'],axis=1,inplace=True)
df.info()
```

```python
from sklearn.model_selection import train_test_split
x_trs,x_tes,y_trs,y_tes = train_test_split(df,y_tr,random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
lr = [0.05,0.1,0.25,0.5,0.75,1]
for i in lr:
  gb=GradientBoostingClassifier(n_estimators=2, learning_rate=i , max_features=2, max_depth=2, random_state= 0)
  gb.fit(x_trs,y_trs)
  print("Learning Rate:\n",i)
  print("Training {0:.3f}".format(gb.score(x_trs, y_trs)))
  print("Validate {0:.3f}".format(gb.score(x_tes, y_tes)))
```
